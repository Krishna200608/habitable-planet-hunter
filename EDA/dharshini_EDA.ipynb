```python
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
plt.hist(df['G3'], bins=20)
plt.title("Distribution of Final Grades (G3)")
plt.xlabel("Final Grade")
plt.ylabel("Number of Students")
plt.show()
```

****The final grades (G3) show a slightly left-skewed distribution, indicating more students scoring in the mid to higher range.****

```python
df.groupby('sex')['G3'].mean()
df.groupby('sex')['G3'].mean().plot(kind='bar')
plt.title("Average Final Grade by Gender")
plt.ylabel("Average G3")
plt.show()
```

**Gender-wise Performance Analysis**
The average final grade (G3) differs slightly across genders.
Male students show a marginally higher mean final grade compared to female students in this dataset.
This observation is dataset-specific and may vary across different populations.

```python
df.groupby('studytime')['G3'].mean()
df.groupby('studytime')['G3'].mean().plot(kind='line', marker='o')
plt.title("Study Time vs Final Grade")
plt.xlabel("Study Time")
plt.ylabel("Average G3")
plt.show()
```

Increased study time is positively correlated with higher final grades.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns
numeric_df = df.select_dtypes(include=['int64', 'float64'])

plt.figure(figsize=(12,8))
sns.heatmap(numeric_df.corr(), cmap='coolwarm')
plt.title("Correlation Heatmap (Numeric Features)")
plt.show()
```

**Correlation Analysis**
The correlation heatmap was generated using only numeric features.
Strong positive correlations are observed between G1, G2, and G3, indicating that students who perform well in earlier assessments tend to maintain their performance in the final grade.
